```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
```

```python
well_data=pd.read_csv("/content/log.csv")
well_data
```

```python
list_column = ["RHOB", "SP","RILD", "DPOR","GR"]
color = [ "green", "red", "blue","purple","black"]
plt.figure(figsize=(11,7))
for i in range(len(list_column)):
  plt.subplot(1,5,i+1)
  # Give exception for RDEP for semilog plot
  if list_column[i]=="RILD":
    plt.semilogx(well_data[list_column[i]], well_data["Depth"], color=color[i])
  else:
    # Others are normal plot
    plt.plot(well_data[list_column[i]], well_data["Depth"], color=color[i])
  
  plt.ylim(3500,1500)
  plt.title(list_column[i])
  plt.grid()

plt.tight_layout(1.5)
plt.savefig('well_log plot.png')
plt.show()
```
